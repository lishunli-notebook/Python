```python
def taobao_search(pages = 2):
    
    import time as TM
    page = 1  #页数
    index = 0  #爬取的个数
    column = ['price', 'buy_person', 'content', 'store']
    df = pd.DataFrame(columns=column)
    
    while page <= pages:
        
        i = 1  #当前页的个数
        key = True

        while key:
            try:
                #第一页第一个内容
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[1]/div[1]/strong
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[1]/div[2]
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[2]
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[3]/div[1]/a/span[2]

                #第一页第二个内容
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[2]/div[2]/div[1]/div[1]/strong
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[2]/div[2]/div[1]/div[2]
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[2]/div[2]/div[2]
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[2]/div[2]/div[3]/div[1]/a/span[2]


                #第二页第一个内容
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[1]/div[1]/strong
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[1]/div[2]
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[2]
                #//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[1]/div[2]/div[3]/div[1]/a/span[2]

                url = '//*[@id="mainsrp-itemlist"]/div/div/div[1]/div[%s]'%(i) #第几个

                price_url = '/div[2]/div[1]/div[1]/strong'
                buy_person_url = '/div[2]/div[1]/div[2]'
                content_url = '/div[2]/div[2]'
                store_url = '/div[2]/div[3]/div[1]/a/span[2]'

                price = taobao.find_element_by_xpath(url+price_url).text
                buy_person = taobao.find_element_by_xpath(url+buy_person_url).text
                content = taobao.find_element_by_xpath(url+content_url).text
                store = taobao.find_element_by_xpath(url+store_url).text

                df.loc[index,:] = [price, buy_person, content, store]

                os.system('say "%s"'%(index))
                index += 1
                i += 1
                TM.sleep(np.random.rand()+1)
                

            except:
                os.system('say "嘟嘟"')
                print('已找到%s页的内容!'%(page))
                key = False
                page += 1 #更新页数
                taobao.find_element_by_link_text('下一页').click()
                
                
    taobao.quit()         
    return df

        
        
        
        
    
```
